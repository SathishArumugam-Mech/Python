# Create a dataframe, save it as csv file, set index as Random

```python
import numpy as np
import pandas as pd

#np arrays
random = np.random.randint(low = 0, high =100, size = 10)
name = np.random.choice(['A', 'B', 'C', 'D', 'E'],size =  10)
choice = np.random.choice([10,11,13,12,14],size = 10)

#creating dictionary
dict1 = {
    'Random':random, 
    'Name':name,
    'Choice':choice
}

#creating dataframe
df = pd.DataFrame(dict1)
print(df,"\n")
df.to_csv("Pandas_DataFrame")
#df = df.set_index('Random')   # This will Create new object df
df.set_index('Random',inplace = True)
print(df)

# Sorting Index
#df.sort_index(ascending=True)

# Sorting Values by Column wise
#df.sort_values(by='Choice', ascending = True)
```

# Create a dataframe,merge it with new column

```python
random = np.random.randint(low = 0, high =100, size = 10)
name = ["Dawid Malan","Babar Azam","KL Rahul","Aaron Finch","Rassie van der Dussen","Glenn Maxwell","Virat Kohli","olin Munro","Tim Seifert","Hazratullah Zazai"]
choice = np.random.choice([10,11,13,12,14],size = 10)

#creating dictionary
dict1 = {
    'Random':random, 
    'Name':name,
    'Choice':choice
}

#creating dataframe-1
df1 = pd.DataFrame(dict1)
print(df1,"\n")

choice1 = np.random.choice([100,200,300,400,500],size = 10)
name = ["Dawid Malan","Babar Azam","KL Rahul","Aaron Finch","Rassie van der Dussen","Glenn Maxwell","Virat Kohli","olin Munro","Tim Seifert","Hazratullah Zazai"]
dict2 = {'Name':name,'Choice1':choice1}
#creating dataframe-2
df2 = pd.DataFrame(dict2)
print(df2,"\n")
pd.merge(df1,df2, on='Name', how ='outer')
```
